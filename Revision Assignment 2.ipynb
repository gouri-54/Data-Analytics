```python
transactions = {
    "T1": ["Milk", "Bread"],
    "T2": ["Bread", "Diaper", "Beer", "Eggs"],
    "T3": ["Milk", "Diaper", "Beer", "Cola"],
    "T4": ["Bread", "Milk", "Diaper", "Beer"],
    "T5": ["Bread", "Milk", "Diaper", "Cola"]
}
print(transactions)
```

```python
1. How many transactions are there?
```

```python
len(transactions)
```

```python
2. Print all transaction IDs
```

```python
for t_id in transactions:
    print(t_id)
```

```python
3. Print items in transaction T3
```

```python
transactions["T3"]
```

```python
4. Create a single list of all items
```

```python
all_items = []

for t in transactions:
    for item in transactions[t]:
        all_items.append(item)

all_items

```

```python
5. How many times does Milk appear?
```

```python
all_items.count("Milk")
```

```python
6. How many transactions contain Bread?
```

```python

count=0
for items in transactions.values():
    if "Bread" in items:
        count += 1

count
```

```python
7.Find all unique items using a set
```

```python

unique_items = set(all_items)
unique_items
```

```python
8.Create a sorted list of unique items
```

```python
sorted_unique_items = sorted(unique_items)
sorted_unique_items
```

```python
9.Count how many times each item appears
```

```python
item_count = {}

for item in all_items:
    if item in item_count:
        item_count[item] += 1
    else:
        item_count[item] = 1

item_count
```

```python
10.Find the item that appears the most
```

```python
most_common_item = max(item_count, key=item_count.get)
most_common_item
```
